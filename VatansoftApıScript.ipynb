```python
#Gerekli Modüller
import requests  # Bu modül ile web üzerindeki isteklerinizi yöneteceksiniz. Mesela bu modül ile API entpointlerine PUT, DELETE, POST gibi istekler atabilirsiniz.
import json      # API'lerden Json verilerini yazdırmak için yaygın olarak kullanılır.
import time      # API çağrıları arasında gecikme oluşturmak için kullanılır.Programın çok fazla isteği çok hızlı bir şekilde yapmasını önler.
```

```python
url = "https://random.dog/woof.json"                    # API Url'i
```

```python

large_size=0    # 1050000 byte dan büyük olan veriler için 
small_size=0    # 1050000 byte dan küçük olan veriler için 

for i in range (100):                                   # 100 adet istek (request) göndermek için bir döngü oluşturuyoruz.    
    response = requests.get(url)                        # Get methodu ile API'den verileri çekmek için istek gönderir                   
    time.sleep(1)                                       # API istekleri arasında 2 saniyelik gecikme
    
    if response.status_code==200:
        data=response.json()                            # API'den verileri Json formatında almak için
        response_size=data.get("fileSizeBytes")         # Her bir isteğin boyutunu içeren "fileSizeBytes" başlığının (headers) değerini almak için  
        #print(response_size)           
        
        if response_size>1050000:                                 
            large_size +=1
            
        else:
            small_size +=1
            
    else:
        print(f"Error:{response.status_code}")

print(f"1050000 byte'dan büyük olanların sayısı {large_size} ") 
print(f"1050000 byte'dan küçük olanların sayısı {small_size} ")
        
```
