```python
#Gerekli Modller
import requests  # Bu modl ile web zerindeki isteklerinizi yneteceksiniz. Mesela bu modl ile API entpointlerine PUT, DELETE, POST gibi istekler atabilirsiniz.
import json      # API'lerden Json verilerini yazdrmak iin yaygn olarak kullanlr.
import time      # API arlar arasnda gecikme oluturmak iin kullanlr.Programn ok fazla istei ok hzl bir ekilde yapmasn nler.
```

```python
url = "https://random.dog/woof.json"                    # API Url'i
```

```python

large_size=0    # 1050000 byte dan byk olan veriler iin 
small_size=0    # 1050000 byte dan kk olan veriler iin 

for i in range (100):                                   # 100 adet istek (request) gndermek iin bir dng oluturuyoruz.    
    response = requests.get(url)                        # Get methodu ile API'den verileri ekmek iin istek gnderir                   
    time.sleep(1)                                       # API istekleri arasnda 2 saniyelik gecikme
    
    if response.status_code==200:
        data=response.json()                            # API'den verileri Json formatnda almak iin
        response_size=data.get("fileSizeBytes")         # Her bir istein boyutunu ieren "fileSizeBytes" balnn (headers) deerini almak iin  
        #print(response_size)           
        
        if response_size>1050000:                                 
            large_size +=1
            
        else:
            small_size +=1
            
    else:
        print(f"Error:{response.status_code}")

print(f"1050000 byte'dan byk olanlarn says {large_size} ") 
print(f"1050000 byte'dan kk olanlarn says {small_size} ")
        
```
